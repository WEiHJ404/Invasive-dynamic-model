```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df =pd.read_excel('/Users/weihj/Desktop//2022/COVID_article/181/suzhou/suzhou_b_4.xlsx',sheet_name='M')
plt.rcParams['font.sans-serif'] = 'Times New Roman'
plt.rcParams.update({'font.size': 20})
fig,ax = plt.subplots(figsize=(18,8))
for i in range(1,11):
    ax.plot(df['Date'][1:],df.iloc[:,i][1:],label=df.iloc[:,i][0])
plt.legend()
dt1 = datetime.datetime(2022,2,10)
dt2 = datetime.datetime(2022,5,31)
plt.xlim(dt1,dt2)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.text(0.005, 0.95, s="M_4 Suzhou City", transform=ax.transAxes,size=26)
# plt.savefig('/Users/weihj/Desktop//2022/COVID_article/181/suzhou/suzhou_M.pdf', dpi=300, bbox_inches = 'tight')
```
